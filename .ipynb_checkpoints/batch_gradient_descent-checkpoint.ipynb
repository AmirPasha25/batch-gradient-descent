# BATCH GRADIENT DESCENT

```python
# Gradient: Gradient is the partial derivative or vector of partial derivaties of a loss function with respect to its parameters.
# Why partial derivative: Partial derivatives tell us how the loss function changes or rate of change of a function
# with respect to each individual parameter (like weight m or bias b) while keeping all other parameters constant.
# Partial derivatives tell us how the loss function changes with respect to each parameter individually
```

```python
# Gradient Descent: 
    # Gradient descent is a iterative optimization algorithm for finding global minimum of function.
    # Gradient Descent Formula
    # θ = θ − α * ∂J/∂θ
```

```python
# Types of gradient descent
    # Batch gradient descent : 
        # Batch Gradient Descent computes the gradient using the entire dataset before updating the weights.
        # It is stable and accurate but becomes slow for very large datasets.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# when you expland this..
# J(m,b)=(y−(mx+b))2
# J=(x^2)m^2+b^2+2xbm+(constant). This is nothing but a quadratic equation
# linear regression with MSE loss always produces a convex paraboloid because:
    # The loss is a sum of squared linear terms
    # Squares are always convex
    # Sum of convex functions is convex
    # The Hessian is always positive-semi-definite
# So for linear regression, the loss surface is always a convex bowl.

m = np.linspace(-5, 5, 200)
b = np.linspace(-5, 5, 200)

M, B = np.meshgrid(m, b)
J = M**2 + B**2

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(M, B, J)

ax.set_title("3D Convex Bowl for Gradient Descent")
ax.set_xlabel("m")
ax.set_ylabel("b")
ax.set_zlabel("Loss J")

plt.show()
```

```python
# Metrics
mse = mean_squared_error(y_test, predict)
r2 = r2_score(y_test, predict)
slope = lr.coef_
intercept = lr.intercept_
print("mse:", mse)
print("r2", r2)
print("slope (m):", lr.coef_)
print("intercept", lr.intercept_)
```

```python
# Starting everything from scratch.

class BatchGradientDescent:
    def __init__(self, X, y, alpha = 0.0001):
        self.alpha = alpha
        self.X = X
        self.y = y
        self.m = np.array([-140]) # this is random value
        self.b = 15.0 # this is random value

    def calculation(self):
        while True:
            y_hat = np.dot(self.X, self.m) + self.b
            error = self.y - y_hat
            dm = -2 * np.dot(self.X.T, error)
            step_size_m = self.alpha * dm    
            db = -2 * np.sum(error)
            step_size_b = self.alpha * db
            previous_m = self.m.copy()
            previous_b = self.b
            self.m = (self.m-step_size_m)
            self.b = (self.b-step_size_b)
            
            if abs(self.m-previous_m)<=0.000001 and abs(self.b-previous_b) <=0.000001:
                break
        print(self.m, self.b)
        
onegradient = BatchGradientDescent(X_train, y_train)
onegradient.calculation()
```

# CODE MULTI VARIATE

```python
# importing necessary packages
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

X,y = make_regression(n_samples = 100, n_features = 10, n_targets = 1, noise = 80, random_state = 42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

# OLS of Sklearn
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print("slope: ", lr.coef_)
print("intercept: ", lr.intercept_)
print("r2 score: ", r2_score(y_pred, y_test))
```

```python
# slope = [m1, m2, m3, .... m10]
# I made this using loop not vectorized.

slope = np.zeros(X_train.shape[1])
intercept = 0.0
learning_rate = 0.001

for iteration in range(1000):    
    # y_hat = m1x1+m2x2+....+b
    y_hat = np.dot(X_train, slope)+intercept
    # error = y_train which is actual value - predicted value
    error = y_train - y_hat

    # Previous slope
    previous_slope = slope.copy()
    # Previous intercept
    previous_intercept = intercept
    
    for i in range(X_train.shape[1]):
        # derivative_m = -2 * Σ (X * error)
        d_slope = -2/len(X_train) * np.sum(X_train[:, i] * error)
        # step size = learning rate * derivative of slope
        step_size_m = learning_rate * d_slope
        # updated slope => slope = slope - step size
        slope[i] = slope[i] - step_size_m
    
    # derivative_b = -2 * Σ (error)
    d_intercept = -2/len(X_train) * np.sum(error)
    # step size = learning rate * derivative of intercept
    step_size_b = learning_rate * d_intercept
    # updated intercept => intercept = intercept - step size
    intercept = intercept - step_size_b

    # Listen carefully..
    # result = current [m1, m2, m3, .... m10] - previous[m1, m2, m3, .... m10] 
    # max(result) > 0.0001
    # don't break
    if np.max(np.abs(slope-previous_slope)) < 0.0001 and abs(intercept-previous_intercept) < 0.0001:
        break


# prediction 
y_predict = np.dot(X_test, slope) + intercept
print(slope)
print(intercept)
print("r2 score: ", r2_score(y_test, y_predict))
```

```python
# 1. I started gradient descent from a random point on a function that has multiple local minima and one global minimum. 
    # My starting point was in the basin of a local minimum, and gradient descent moved downhill until it reached that local minimum. 
    # How does gradient descent know that there are other minima it should explore in order to find the global minimum?

# Answer: 
# Gradient descent is inherently a local optimization method, meaning it only uses the slope at the current point to decide which direction to move.
# If it starts in the basin of a local minimum, it will continue descending until it reaches that local minimum and will not be aware of other minima elsewhere in the function.
# It has no built-in mechanism to “look ahead” or explore other valleys.
# To find the global minimum in a landscape with multiple minima, additional strategies are required.
# Common solutions include performing multiple random restarts from different initial points to increase the chance of landing in the basin of the global minimum.
# Using stochastic methods or adding noise to the updates can help escape shallow local minima.
# Employing optimizers with momentum or adaptive learning rates can sometimes help overcome small local traps.
# In essence, gradient descent alone follows the slope locally, and finding the global minimum requires explicit techniques for global exploration.


from PIL import Image

image = Image.open(r"C:\Users\amirp\Desktop\Project_11_Gradient_Descent\Local_Global_Minima.png")
image.show()


# Explanation
# Your sketch represents a loss function with many ups and downs.
    # Gradient Descent (GD) tries to go downhill to reduce loss, but the problem is:
        # The surface has many local minima.
        # Only one point is the global minimum (lowest loss overall).
    # A simple GD always finds the local minima of that basin.
    # Wherever you start, GD will follow the slope and settle in the closest dip
    # This dip might not be the best (global minimum).
    # So GD is greedy — it only looks at the slope near itself.

# How to overcome it?
    # Starting random points / restarting helps find global minima.
    # If you run GD from different random starting points:
        # you fall into different dips, increasing chances that one run reaches the global minimum.
            # This technique is called:
                # ➡️ Random Restarts
                # ➡️ Multi-start Gradient Descent
    # Use stochastic GD:
        # Stochastic Gradient Descent (SGD) adds noise because it uses batches, not full data.
        # That noise can:
            # kick you out of a shallow local minima
            # push you towards a deeper/global minimum.
        # So SGD is more likely to find a better minimum than normal GD
```
